```python
"""
Блок 2. Python

"""

# ---------------------------------------------------------------------------
# Задание 1. Изоморфизмы
# ---------------------------------------------------------------------------


def is_isomorphic(s: str, t: str) -> bool:
    if len(s) != len(t):
        return False
    forward: dict[str, str] = {}  
    backward: dict[str, str] = {} 
    for cs, ct in zip(s, t):
        if cs in forward:
            if forward[cs] != ct:
                return False
        elif ct in backward:
            return False
        else:
            forward[cs] = ct
            backward[ct] = cs
    return True

def is_isomorphic_zip(s: str, t: str) -> bool:
    return len(s) == len(t) and len(set(s)) == len(set(t)) == len(set(zip(s, t)))


# ---------------------------------------------------------------------------
# Задание 2. Натуральная последовательность
# ---------------------------------------------------------------------------


def missing_number(nums: list[int]) -> int:
    n = len(nums) + 1
    return n * (n + 1) // 2 - sum(nums)


def missing_number_xor(nums: list[int]) -> int:
    n = len(nums) + 1
    result = 0
    for i in range(1, n + 1):
        result ^= i
    for x in nums:
        result ^= x
    return result


def missing_number_sorted(nums: list[int]) -> int:
    lo, hi = 0, len(nums) - 1
    while lo <= hi:
        mid = (lo + hi) // 2
        if nums[mid] == mid + 1:  
            lo = mid + 1
        else:                       
            hi = mid - 1
    return lo + 1


# ---------------------------------------------------------------------------
# Задание 3. Факторизация
# ---------------------------------------------------------------------------


def prime_factors(n: int) -> list[int]:
    if n < 2:
        return []
    factors: list[int] = []
    while n % 2 == 0:
        factors.append(2)
        n //= 2
    divisor = 3
    while divisor * divisor <= n:
        while n % divisor == 0:
            factors.append(divisor)
            n //= divisor
        divisor += 2
    if n > 1:
        factors.append(n)
    return factors


# ---------------------------------------------------------------------------
# Проверка
# ---------------------------------------------------------------------------

if __name__ == "__main__":
    # 1
    assert is_isomorphic("paper", "title") is True
    assert is_isomorphic("egg", "add") is True
    assert is_isomorphic("foo", "bar") is False
    assert is_isomorphic("badc", "baba") is False   
    assert is_isomorphic("ab", "abc") is False      
    assert is_isomorphic("", "") is True
    for f in (is_isomorphic, is_isomorphic_zip):
        assert f("paper", "title") and not f("badc", "baba")
    print("Задание 1 (изоморфизмы): ок")
    print("  is_isomorphic('paper', 'title') ->", is_isomorphic("paper", "title"))

    # 2
    assert missing_number([1, 2, 3, 4, 5, 6, 8, 9, 10, 11]) == 7
    assert missing_number([2, 3]) == 1
    assert missing_number([1, 2]) == 3
    for f in (missing_number, missing_number_xor, missing_number_sorted):
        assert f([1, 2, 3, 4, 5, 6, 8, 9, 10, 11]) == 7
        assert f([2, 3, 4]) == 1
        assert f([1, 2, 3]) == 4
    print("Задание 2 (пропущенное число): ок")
    print("  missing_number([1,2,3,4,5,6,8,9,10,11]) ->",
          missing_number([1, 2, 3, 4, 5, 6, 8, 9, 10, 11]))

    # 3
    from math import prod
    assert prime_factors(56) == [2, 2, 2, 7]
    assert prime_factors(97) == [97]
    assert prime_factors(1) == []
    assert prime_factors(2) == [2]
    for n in range(2, 2000):
        f = prime_factors(n)
        assert prod(f) == n
    print("Задание 3 (факторизация): ок")
    print("  prime_factors(56) ->", prime_factors(56))
```
